```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import os
```

```python
url = "https://tuoitre.vn"
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
html = urlopen(req)
soup = BeautifulSoup(html, 'html')
```

```python
title_, link_ = [], []

for c in ['title-focus-title', 'title-news-list-focus', 'title-box-photo', 'title-news', 'title-box-wc-2018']:
    for i in soup.find_all(class_=c):
        title_.append(i.text)
        link_.append('https://tuoitre.vn' + i.find_next('a').get('href'))
for i in soup.find_all('h2', class_='title-name'):
    title_.append(i.text)
    link_.append('https://tuoitre.vn' + i.find_next('a').get('href'))
for i in soup.find_all('h3', class_='title-name'):
    title_.append(i.text)
    link_.append('https://tuoitre.vn' + i.find_next('a').get('href'))
for i in soup.find_all('h3', class_='name-title'):
    title_.append(i.text)
    link_.append('https://tuoitre.vn' + i.find_next('a').get('href'))
for i in soup.find_all('a', class_='title-name'):
    title_.append(i.text)
    link_.append('https://tuoitre.vn' + i.get('href'))
```
